```python
# prompt: read my file solicitacoescredito.csv inside MyDrive

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/coisas do colab/solicitacoescredito.csv')

df.describe()
```

```python
i = df[(df.status == "DocumentacaoReprovada")].index # Classe não relevante para predição
df = df.drop(i)

i = df[(df.status == "EmAnaliseDocumentacao")].index # Classe não relevante para predição
df = df.drop(i)

i = df[(df.status == "AguardandoAprovacao")].index # Removemos essa classe por ausência de exemplos
df = df.drop(i)

df.describe()
```

```python
df['faturamentoBruto'].describe()
```

```python
# i = df[(df.valorSolicitado > 100000000)].index
# df = df.drop(i)

# df.describe()

i = df[((df['valorSolicitado'].quantile(0.75) * 1.5) < df.valorSolicitado)].index
j = df[((df['valorSolicitado'].quantile(0.25) * 0.5) > df.valorSolicitado)].index
df = df.drop(i)
df = df.drop(j)


i = df[((df['faturamentoBruto'].quantile(0.75) * 1.5) < df.faturamentoBruto)].index
j = df[((df['faturamentoBruto'].quantile(0.25) * 0.5) > df.faturamentoBruto)].index
df = df.drop(i)
df = df.drop(j)


i = df[((df['valorAprovado'].quantile(0.75) * 1.5) < df.valorAprovado)].index
j = df[((df['valorAprovado'].quantile(0.25) * 0.5) > df.valorAprovado)].index
df = df.drop(i)
df = df.drop(j)


i = df[((df['maiorAtraso'].quantile(0.75) * 1.5) < df.maiorAtraso)].index
j = df[((df['maiorAtraso'].quantile(0.25) * 0.5) > df.maiorAtraso)].index
df = df.drop(i)
df = df.drop(j)

```

```python
# prompt: Add a column that sums this equation from other columns: ativoCirculante + passivoCirculante + totalAtivo + totalPatrimonioLiquido - endividamento - duplicatasAReceber + faturamentoBruto

df['indiceBalanco'] = df['ativoCirculante'] + df['passivoCirculante'] + df['totalAtivo'] + df['totalPatrimonioLiquido'] - df['endividamento'] - df['duplicatasAReceber'] + df['faturamentoBruto']

i = df[((df['indiceBalanco'].quantile(0.75) * 1.5) < df.indiceBalanco)].index
j = df[((df['indiceBalanco'].quantile(0.25) * 0.5) > df.indiceBalanco)].index
df = df.drop(i)
df = df.drop(j)
```

```python
X = df[[
    'valorSolicitado',
    'maiorAtraso',
    'faturamentoBruto',
    'scorePontualidade',
    'percentualRisco',
    'valorAprovado',
    'indiceBalanco'
    ]]

Y = df['status']
```

```python
#Vamos explorar um pouco uma matrix de correlação

import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = X.corr()

fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(data=correlation_matrix, annot=True, linewidths=0.5, ax=ax)
```

```python
X.plot.scatter('valorSolicitado', 'indiceBalanco')

df.fillna(0.0, inplace=True)

X = df[[
    'valorSolicitado', 'indiceBalanco'
    ]]

Y = df['status']

df['status'].value_counts()
```

```python
# Separamos 20% para o teste
from sklearn.model_selection import train_test_split

X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size=0.2)

print(X_treino.shape)
print(X_teste.shape)
print(Y_treino.shape)
print(Y_teste.shape)
```

```python
# prompt: balance my training set

from imblearn.over_sampling import SMOTE

# Set k_neighbors to a value less than the smallest class size
smote = SMOTE(random_state=42, k_neighbors=1) # k_neighbors MUST be less than the smallest class size
X_treino_balanceado, Y_treino_balanceado = smote.fit_resample(X_treino, Y_treino)

print(Y_treino_balanceado.value_counts())
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X_treino_balanceado)
y_kmeans = kmeans.predict(X_treino_balanceado)
```

```python
plt.scatter(X_treino_balanceado['valorSolicitado'], X_treino_balanceado['indiceBalanco'], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
df_groups = pd.DataFrame(X_treino_balanceado, columns=['valorSolicitado', 'indiceBalanco'])
df_groups['cluster'] = y_kmeans

df_groups.head()
```

```python
# Merge Y_treino with df_groups based on index
merged_df = Y_treino_balanceado.to_frame().merge(df_groups, left_index=True, right_index=True)

merged_df.groupby('cluster')['status'].value_counts()
```

```python
merged_df.groupby('cluster')['status'].value_counts(normalize=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Create a KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Fit the model on the training data
knn.fit(X_treino_balanceado, Y_treino_balanceado)

# Predict on the test data
y_pred = knn.predict(X_teste)

# Calculate the accuracy
accuracy = accuracy_score(Y_teste, y_pred)
print(f"Accuracy: {accuracy}")

# Predict clusters for the test data
y_kmeans_test = kmeans.predict(X_teste)

# Create a DataFrame for the test data with the predicted clusters
df_groups_test = pd.DataFrame(X_teste, columns=['valorSolicitado', 'indiceBalanco'])
df_groups_test['cluster'] = y_kmeans_test

# Merge Y_teste with df_groups_test based on index
merged_df_test = Y_teste.to_frame().merge(df_groups_test, left_index=True, right_index=True)

# Display the grouped values for the test data
print(merged_df_test.groupby('cluster')['status'].value_counts())
print(merged_df_test.groupby('cluster')['status'].value_counts(normalize=True))
```

```python
# prompt: show me the F1 score of the model above

from sklearn.metrics import f1_score

# Calculate the F1 score
f1 = f1_score(Y_teste, y_pred, average='weighted')  # Use 'weighted' for multi-class problems
print(f"F1 Score: {f1}")
```

```python
# prompt: Show me a confusion matrix with the relation between prediction and actual

from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the confusion matrix
cm = confusion_matrix(Y_teste, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=knn.classes_, yticklabels=knn.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
df = pd.read_csv('/content/drive/MyDrive/coisas do colab/solicitacoescredito.csv')

df.describe()
```

```python
i = df[(df.status == "DocumentacaoReprovada")].index
df = df.drop(i)

i = df[(df.status == "EmAnaliseDocumentacao")].index
df = df.drop(i)

i = df[(df.status == "ReprovadoComite")].index
df = df.drop(i)

i = df[(df.status == "ReprovadoAnalista")].index
df = df.drop(i)

i = df[((df['valorSolicitado'].quantile(0.75) * 1.5) < df.valorSolicitado)].index
j = df[((df['valorSolicitado'].quantile(0.25) * 0.5) > df.valorSolicitado)].index
df = df.drop(i)
df = df.drop(j)

i = df[((df['faturamentoBruto'].quantile(0.75) * 1.5) < df.faturamentoBruto)].index
j = df[((df['faturamentoBruto'].quantile(0.25) * 0.5) > df.faturamentoBruto)].index
df = df.drop(i)
df = df.drop(j)

i = df[((df['valorAprovado'].quantile(0.75) * 1.5) < df.valorAprovado)].index
j = df[((df['valorAprovado'].quantile(0.25) * 0.5) > df.valorAprovado)].index
df = df.drop(i)
df = df.drop(j)


i = df[((df['maiorAtraso'].quantile(0.75) * 1.5) < df.maiorAtraso)].index
j = df[((df['maiorAtraso'].quantile(0.25) * 0.5) > df.maiorAtraso)].index
df = df.drop(i)
df = df.drop(j)

df.describe()
```

```python
X2 = df[['maiorAtraso',
        'margemBrutaAcumulada',
        'percentualProtestos',
        'prazoMedioRecebimentoVendas',
        'titulosEmAberto',
        'valorSolicitado',
        'ativoCirculante',
        'passivoCirculante',
        'totalAtivo',
        'totalPatrimonioLiquido',
        'endividamento',
        'duplicatasAReceber',
        'estoque',
        'faturamentoBruto',
        'margemBruta',
        'periodoDemonstrativoEmMeses',
        'custos',
        'capitalSocial',
        'scorePontualidade',
        'limiteEmpresaAnaliseCredito'
      ]]

Y2 = df[['valorAprovado']]
```

```python
# prompt: Change all NaN values into 0.0

X2.fillna(0.0, inplace=True)
Y2.fillna(0.0, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, Y_treino, Y_teste = train_test_split(X2, Y2, test_size=0.2)

print(X_treino.shape)
print(X_teste.shape)
print(Y_treino.shape)
print(Y_teste.shape)
```

```python
# prompt: Use RandomForestRegressor to train a model using X_treino and Y_treino

from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regressor model
rf_model = RandomForestRegressor()

# Train the model
rf_model.fit(X_treino, Y_treino.values.ravel())
```

```python
# prompt: Now test the model using X_teste and give me r2 and other factors to check accuracy using Y_teste

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

# Make predictions on the test set
y_pred = rf_model.predict(X_teste)

# Calculate R-squared
r2 = r2_score(Y_teste, y_pred)
print(f"R-squared: {r2}")

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(Y_teste, y_pred)
print(f"Mean Squared Error: {mse}")

# Calculate Root Mean Squared Error (RMSE)
rmse = mean_squared_error(Y_teste, y_pred, squared=False)
print(f"Root Mean Squared Error: {rmse}")

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(Y_teste, y_pred)
print(f"Mean Absolute Error: {mae}")
```
